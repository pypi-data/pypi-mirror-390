```python
# --- Imports ---
import torch
import ipywidgets as widgets
from IPython.display import display, HTML
import pickle
import numpy as np

from elektronn.ensemble_predict import (
    read_xyz_file_corrected,
    get_iso_permuted_molecule,
    ElektroNN_Ensemble,
    basisfunction_params,
)
from elektronn.visualization_utils import LoadModels, model_kwargs
from elektronn.utils import coeff_unperm_gau2grid_density_kdtree_ml_only, generate_grid

# --- Settings ---
num_models = 5
model_path = "/home/luisakaermer/elektronn-pip/src/elektronn/modelparams/04-20-33/"
map_location = "cpu"
version = "2.0"
```

```python
# --- Imports ---

xyz_file = "/home/luisakaermer/elektronn-pip/template/benzene.xyz"
```

```python
# ===============================
# 2. Load Models
# ===============================
loader_specified = LoadModels(
    models=[2],  # load model 2
    num_models=num_models,
    model_path=model_path,
    model_kwargs=model_kwargs,
    map_location=map_location,
    all_models=True
)
loader_specified.load()
models_specified = loader_specified.models
print(f"Loaded {len(models_specified)} models")

# ===============================
# 3. Load Molecule
# ===============================
mol_data = read_xyz_file_corrected(xyz_file)

encoded = get_iso_permuted_molecule(
    x=mol_data[1],
    pos_orig=mol_data[2],
    atomic_numbers=mol_data[0],
    exps=mol_data[3],
    norms=mol_data[4],
    filename=xyz_file
)
```

```python
# ===============================
# 4. Predict Electron Density
# ===============================
pred = ElektroNN_Ensemble(
    encoded,
    [models_specified[2]],
    map_location=map_location
)
encoded.full_c = pred
print("ML prediction complete")
```

```python
mol = read_xyz_file_corrected("/home/luisakaermer/elektronn-pip/template/benzene.xyz")
encoded = get_iso_permuted_molecule(x = mol[1],pos_orig=mol[2],atomic_numbers=mol[0],exps=mol[3],norms=mol[4],filename="/home/luisakaermer/elektronn-pip/template/benzene.xyz")
pred = ElektroNN_Ensemble(encoded, [models_specified[2]], map_location=map_location)
encoded.full_c = pred
```

```python
# ===============================
# 5. Generate 3D Grid
# ===============================
x, y, z, vol, x_spacing, y_spacing, z_spacing = generate_grid(
    encoded,
    spacing=0.2,
    buffer=2.0
)
print("3D grid generated")

# ===============================
# 6. Compute ML Density on Grid
# ===============================
Rs = [(14, 0), (14, 1), (5, 2), (4, 3), (2, 4)]

ml = coeff_unperm_gau2grid_density_kdtree_ml_only(
    x.flatten().astype(np.float64),
    y.flatten().astype(np.float64),
    z.flatten().astype(np.float64),
    encoded,
    pred,  # keep as torch.Tensor
    Rs
)
print("ML density computed on grid")

# ===============================
# 7. Save Density to Cube File
# ===============================
cube_file = "/home/luisakaermer/elektronn-pip/template/predicted_density.cube"
write_cube_file(cube_file, encoded.atomic_numbers, encoded.pos_orig, x, y, z, ml)
print(f"Density saved to {cube_file}")
```
